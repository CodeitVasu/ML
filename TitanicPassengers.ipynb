```python
df = pd.read_csv('titanic.csv')
df
```

```python
input=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
input
```

```python
target = input['Survived']
input = input.drop(['Survived'],axis='columns')
input
```

```python
input.info()
```

```python
input
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
input['Sex']=le.fit_transform(input['Sex'])

input
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(input,target,test_size=0.10)
```

```python
model.fit(x_train,y_train)
```

```python
model.score(x_test,y_test)
```

```python
model.predict([[3,1,7.25]])
```
